```python
target = data["ABSTRACT"]
print(target[:5])
```

```python
#Tokenize the data
token_data = [nltk.word_tokenize(word) for word in target]
token_data[:4]
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words("english"))
token_data_1 = []
for words in token_data:
    l = []
    for w in words:
        if w not in stop:
            l.append(w)
    token_data_1.append(l)
token_data_1[:4]
```

```python
import string
punct = set(string.punctuation)
token_data_2 = []
for words in token_data_1:
    l = []
    for w in words:
        if(w not in punct):
            l.append(w)
    token_data_2.append(l)
token_data_2[:4]
```

```python
preprocess_data = []
for i in token_data_2:
    preprocess_data.append(" ".join(i))
preprocess_data[:4]
```

```python
tfidf = TfidfVectorizer(stop_words = "english",max_df = .80,max_features=5000,use_idf = False,norm = None)
tf_vectors = tfidf.fit_transform(preprocess_data)
```

```python
tfidf.get_feature_names()
```

```python
tf_vectors.A
```

```python
print(tf_vectors)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(tf_vectors,data["LABEL"],test_size =0.25, random_state = 42)
```

```python
model = MultinomialNB().fit(X_train, y_train)
model
```

```python
prediction = model.predict(X_test)
prediction
```
